Check for null values in target and predictor columns. If you have null target value, you need to throw that example out. If you have null predictor value, you can drop those rows from the data set or fill them in with something (like mean). Describe what you decided to do with the null data.

```
#Checks for null values in target and predictor columns
print(subset['comb08'].isna().any())
print(subset['year'].isna().any())
print(subset['cylinders'].isna().any())
print(subset['displ'].isna().any())

#Replaces the null values in 'cylinders' and 'displ' because those are the only colunns with missing values
#It also prints True and False to let the programmer know when and where null values have been replaced with the mean
print("before:", subset['cylinders'].isna().any())
cylinders_mean = subset['cylinders'].mean()
subset['cylinders'].fillna(cylinders_mean, inplace = True)
print("after: ", subset['cylinders'].isna().any())

print("before:", subset['displ'].isna().any())
cylinders_mean = subset['displ'].mean()
subset['displ'].fillna(cylinders_mean, inplace = True)
print("after: ", subset['displ'].isna().any())

```

Above I replaced the null values with the average. I decided to do this instead of deleting the null values so I have more data to work with and my data will not be as skewed.

Exercise 3

Write k-nearest neighbors function. It should be able to make mpg("comb08") predictions for new cars based on the year, cylinders, and displacement. You should specify what you want to use as k.

```
#define a function
def knn(car, subset, k):

  #calculate the distance
  subset['distance_to_new'] = numpy.sqrt(
      (car['year'] - subset['year']) ** 2 +
      (car['cylinders'] - subset['cylinders']) ** 2 +
      (car['displ']- subset['displ']) ** 2)
  
  #sort by distance
  sorted_subset = subset.sort_values(['distance_to_new'])
  
  #grab closest neighbors
  print(sorted_subset.iloc[0:k])

  #return mode of k closest neighbors
  print(sorted_subset.iloc[0:k]['comb08'].mean())

  
```

```
#dictionary of values
new_values = {}
new_values['year'] = 1993
new_values['cylinders'] = 4
new_values['displ'] = 2

#K-NN
knn(new_values,subset,2)
```

Exercise 5

Make a copy of the data and normalize the training data using Z-score. Predict the mpg of a particular car using your k-nearest-neighbors function with both the normalized and non-normalized training data. 

```
#copy subset from above
normalized_subset = subset.copy()

#normalized data with z-score
year_mean = normalized_subset['year'].mean()
year_std = normalized_subset['year'].std()
cylinders_mean = normalized_subset['cylinders'].mean()
cylinders_std = normalized_subset['cylinders'].std()
displ_mean = normalized_subset['displ'].mean()
displ_std = normalized_subset['displ'].std()

normalized_subset['year'] = (normalized_subset['year'] - year_mean)/year_std
normalized_subset['cylinders'] = (normalized_subset['cylinders'] - cylinders_mean)/cylinders_std
normalized_subset['displ'] = (normalized_subset['displ'] - displ_mean)/displ_std


new_values2 = {}
new_values2['year'] = (1993 - year_mean)/year_std
new_values2['cylinders'] = (4 - cylinders_mean/cylinders_std)
new_values2['displ'] = (2 - displ_mean/displ_std)

#K-NN
knn(new_values2,normalized_subset,2)
```

Compare your results. Describe and explain the differences in a few sentences

```
knn(new_values,subset,2)
knn(new_values2,normalized_subset,2)
```
